```python
marketing_df["Income"].describe()
```

```python
#outliers for income (to detect customer value)
Q1=marketing_df["Income"].quantile(0.25)
Q2=marketing_df["Income"].quantile(0.50)
Q3=marketing_df["Income"].quantile(0.75)

IQR=Q3-Q1 
upper_bound = Q3+1.5 *IQR
print("Q1: ",Q1)
print("Q2: ",Q2)
print("Q3: ",Q3)
print("IQR:",IQR)
print(upper_bound)
### Income Outlier Analysis

# The IQR method was used to detect potential outliers in the Income column. 
# After inspecting the identified records, the maximum income value (₹258,027) was found to be realistic 
# and not indicative of a data-entry error.

# Since this is a marketing analytics project, high-income customers represent 
# an important customer segment. Removing these observations could negatively
#  affect customer segmentation and business insights.

# Therefore, the income outliers were retained for further analysis.
```

```python
# feature engineering
Current_year=datetime.now().year
marketing_df["Age"]=Current_year-marketing_df["Year_Birth"]
marketing_df[["Year_Birth","Age"]].head()
```

```python
marketing_df["Age"].describe()
#Age Feature Validation:
#The Age feature was created from Year_Birth using the current year. 
# Validation showed ages ranging from 30 to 90 years, 
# which are realistic for the dataset. No invalid or
#  impossible age values were observed.
```

```python
#Customer_Tenure 
marketing_df["Customer_Tenure"] = Current_year-marketing_df["Dt_Customer"].dt.year
marketing_df["Customer_Tenure"]
```

```python

marketing_df["Customer_Tenure"].describe()
### Customer_Tenure Validation

#The Customer_Tenure feature was created by subtracting the customer's
#  joining year from the current year.
#The tenure values range from 12 to 14 years, with an average of 
# approximately 13 years. These values are consistent with the dataset,
#  where customers joined between 2012 and 2014.
#No unrealistic or invalid tenure values were found.
```

```python
# Children
marketing_df["Children"]=marketing_df["Kidhome"]+marketing_df["Teenhome"]
marketing_df["Children"].head()
```

```python
marketing_df["Children"].describe()
### Children Feature Validation

#The `Children` feature was created by combining `Kidhome` and `Teenhome`.

#The values range from 0 to 4 children, with an average of approximately
#  1 child per customer. All values are realistic, and no further cleaning
#  is required.
```

```python
#TOtal Spend
marketing_df["Total_Spend"]=marketing_df["MntWines"]+marketing_df["MntFruits"]+marketing_df["MntMeatProducts"]+marketing_df["MntFishProducts"]+marketing_df["MntSweetProducts"]+marketing_df["MntGoldProds"]
marketing_df["Total_Spend"].head()
```

```python
plt.Figure(figsize=(14,18))
plt.boxplot(marketing_df["Total_Spend"])
plt.title("Box plot for Total Spend Amount")
plt.ylabel("Total_Spend")
plt.show()
```

```python
#outliers for Total_Spend 
Q1=marketing_df["Total_Spend"].quantile(0.25)
Q2=marketing_df["Total_Spend"].quantile(0.50)
Q3=marketing_df["Total_Spend"].quantile(0.75)

IQR=Q3-Q1 
upper_bound2 = Q3+1.5 *IQR
print("Q1: ",Q1)
print("Q2: ",Q2)
print("Q3: ",Q3)
print("IQR:",IQR)
print(upper_bound2)
```

```python
outliers=marketing_df[marketing_df["Total_Spend"]>upper_bound2]
print("Numbers of outliers:", len(outliers))
outliers[["Income","Total_Spend"]].head()
```

```python
### Outlier Analysis – Total_Spend

#Outliers were detected in the `Total_Spend` feature using the IQR method.

#After examining the outlier records, the customers were found 
# to have relatively high incomes along with high spending. 
# These values appear to represent genuine high-value customers 
# rather than data entry errors.

#Therefore, the outliers were retained to preserve 
# valuable business information for customer segmentation and s
# pending analysis.
```

```python
#Total purchase
marketing_df["Total_Purchases"]=marketing_df["NumDealsPurchases"]+marketing_df["NumWebPurchases"]+marketing_df["NumCatalogPurchases"]+marketing_df["NumStorePurchases"]
marketing_df["Total_Purchases"].head()
```

```python
marketing_df["Total_Purchases"].describe()
### Total_Purchases Validation

#The `Total_Purchases` feature was created by summing purchases 
# made through the web, catalog, and store channels.
#The average customer made approximately 13 purchases, 
# with values ranging from 0 to 46. The feature contains realistic 
# values and will be useful for analyzing customer purchasing behavior and
#  segmentation.
```

```python
# Starting EDA
# UNIVARIATE ANALYSIS - we have analysed the dustribution of cutomer income
plt.Figure(figsize=(11,4))
sns.histplot(marketing_df["Income"],bins=50)
plt.title("Distribution of Customer Income")
plt.xlabel("Income")
plt.ylabel("Number of Customers")
plt.show()

### Univariate Analysis – Income

#The histogram shows that most customers fall within the low to middle 
# income ranges, while only a small number belong to the high-income segment.

#The income distribution is positively skewed,
#  indicating that high-income customers are relatively rare.

#**Business Insight:**
#The company should focus its general marketing campaigns on middle-income 
# customers while creating premium offers for high-income customers.
```

```python
# UNIVARIATE ANALYSIS FOR AGE
plt.figure(figsize=(14,6))
sns.histplot(marketing_df["Age"],bins=30)
plt.title("Age Distribution of Customer")
plt.xlabel("Age")
plt.ylabel("number of Customer")
plt.show()
#The histogram shows that customers aged 50–60 years form the largest
#  customer group, while customers aged 75–90 years are comparatively fewer.
#  Therefore, the company should primarily target middle-aged customers
#  in its marketing campaigns, as they represent the majority of the customer 
# base. Separate campaigns can be designed for senior customers based on their
#  preferences.
```

```python
# Total Spend
plt.figure(figsize=(8,8))
sns.histplot(marketing_df["Total_Spend"],bins=30)
plt.title("Distribution of total Expenditure")
plt.xlabel("Total_Spend")
plt.ylabel("Number Of Customers")
plt.show()
#The histogram indicates that most customers have low total spending, 
# while only a small number of customers are high spenders. The company 
# should encourage low-spending customers to purchase more through promotional 
# offers and retain high-spending customers with loyalty programs and
#  exclusive benefits.
```

```python
# Recency
plt.figure(figsize=(10,8))
sns.histplot(marketing_df["Recency"],bins=30)
plt.title("Distribution Of Recency")
plt.xlabel("Recency")
plt.ylabel("Number of Customers")
plt.show()
#The histogram shows that many customers have a high Recency value,
#  meaning they have not made a purchase for a long time.
#  This indicates that a large portion of customers may be inactive.
#  The company should consider running re-engagement campaigns such as 
# discounts, personalized offers, or reminder emails to encourage these 
# customers to make another purchase.
```

```python
# BI VARIATE ANALYSIS
plt.figure(figsize=(12,8))
sns.scatterplot(data=marketing_df , x = "Income", y="Total_Spend", alpha=0.3 , s=20)
plt.title(" income vs Total spend")
plt.xlabel("Income")
plt.ylabel("Total spend")
plt.show()
```

```python
#Observation: The scatter plot shows a moderate positive correlation 
# between Income and Total Spend. Customers with higher incomes 
# generally tend to spend more, although there is some variation in spending 
# behavior.

#Business Insight: The company can target high-income customers with
#  premium products and personalized offers, while encouraging lower-income
#  customers to increase their spending through discounts and promotional
#  campaigns
```
